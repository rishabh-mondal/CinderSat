```python
import torch
from ultralytics import YOLO

# Load YOLOv8 model
model = YOLO('yolov8s.pt')
```

```python
# Load the pretrained MoCo backbone
pretrained_backbone = torch.load('/home/rishabh.mondal/Brick-Kilns-project/albk_rishabh/albk_v2/YOLO_LOCALIZATION/fine_tuned_model.pth')

# Custom function to transfer weights
def transfer_weights(model, pretrained_backbone):
    # Iterate over model's layers and pretrained_backbone's layers and copy weights
    for name, module in model.named_children():
        if name in pretrained_backbone:
            try:
                module.load_state_dict(pretrained_backbone[name])
            except Exception as e:
                print(f"Error loading {name}: {e}")

# Transfer the weights
transfer_weights(model.model, pretrained_backbone)
```
